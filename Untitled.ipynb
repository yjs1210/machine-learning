```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
x_test = pd.read_csv("data/hw1-data/X_test.csv", header = None)
y_test = pd.read_csv("data/hw1-data/y_test.csv", header = None)
x_train = pd.read_csv("data/hw1-data/X_train.csv", header = None)
y_train = pd.read_csv("data/hw1-data/y_train.csv", header = None)
```

```python
def ridge_solution(lambda_,x,y):
    xt = x.transpose()
    return np.matmul(np.linalg.inv((lambda_*np.identity(len(x.columns)) + np.matmul(xt,x))),np.matmul(xt,y))
```

```python
out = []
for lambda_ in range(0,5001):
    w = ridge_solution(lambda_,x_train,y_train)
    out.append(w)
```

```python
ws=[]
for i in range(0,7):
    temp= []
    for j in range(0,5001):
        a = out[j][i]
        temp.append(a)
    ws.append(temp)
```

```python
u, s, v = np.linalg.svd(x_train)
df=[]
for i in range(0,5001):
    temp = sum(s**2 / (s**2 +(i)))
    df.append(temp)
```

```python
%matplotlib inline
fig = plt.figure(figsize=(10,10))
ax1 = plt.gca()
lab = ['Cylinders','Displacement','Horsepower', 'Weight','Acceleration','Year Made','Intercept']
for i in range(0, 7):
        ax1.plot(df, ws[i], label = lab[i])
plt.xlabel("df lambda")
plt.ylabel("Ridge Coefficient")
plt.title("Ridge Coefficient vs. df")
ax1.legend()
```

## 3b)

Degrees of freedom is inversely correlated to the lambda. As we inrease lambda, we are restricting the budget of the coefficients more. So as degrees of freedom decreases, the coefficients start to decrease more. 

We notice thta initially, the weight and year made coefficients have the most marginal predictive impact where as other predictive factos such as horsepower was rather muted. While they still arguably are very important, there is a strong reason to believe that OLS overfits this data since imposing penalty on the model induces rapid decrease in these coefficients. We also see that other features get more emphasized as these two features get penalized. This is probably due to the model initially overfitting and true prdictive powers of the other features are starting to manifest themselves as we penalize the model from using big coefficients and complicated models.

## 3c)

```python
rmse = []
for lambda_ in range(0,51):
    preds = np.sqrt(np.average((y_test - np.matmul(x_test,out[lambda_]))**2))
    rmse.append(preds)
```

```python
plt.plot(rmse)
plt.xlabel('lambda') 
plt.ylabel('rmse')
plt.title('rmse by lambda')
```

RMSE is increasing as we increase lambda. In fact it is minimized with lambda = 0. In other words, using vanilla OLS leads to the best prediction in our test dataset.

```python
x_train_2 = x_train.copy()
x_2_mean = []
x_2_sd = []
```

```python
for i in range(0,6):
    col = x_train_2.iloc[:,i]**2
    x_2_mean.append(np.average(col))
    x_2_sd.append(np.std(col))
    x_train_2['2nd_'+str(i)] = (col-x_2_mean[i])/x_2_sd[i]
```

```python
x_train_3 = x_train_2.copy()
x_3_mean = []
x_3_sd = []
```

```python
for i in range(0,6):
    col = x_train_3.iloc[:,i]**3
    x_3_mean.append(np.average(col))
    x_3_sd.append(np.std(col))
    x_train_3['3rd_'+str(i)] = (col-x_3_mean[i])/x_3_sd[i]
```

```python
out2 = []
out3 = []
for lambda_ in range(0,101):
    w = ridge_solution(lambda_,x_train_2,y_train)
    w3 = ridge_solution(lambda_,x_train_3,y_train)
    out2.append(w)
    out3.append(w3)
```
